```python
# Import Libraries
import twint
import nest_asyncio
nest_asyncio.apply()
import time
import pandas as pd
import os
import re
timestr = time.strftime("%Y%m%d")
```

Configuration to fetch data.

```python
c = twint.Config()   #configuration
c.Limit = 1000       
c.Lang = "en"        # Language
c.Store_csv = True
c.Search = "robin i love you"  # key words to look for.
c.Output = timestr + "_en_robin.csv"
twint.run.Search(c)
```
